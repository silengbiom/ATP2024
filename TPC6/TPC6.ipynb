```python

def criar_turma():
    return []

def inserir_aluno(turma):
    nome = input("Insira o nome do aluno: ")
    id = input("Insira o ID do aluno: ")
    notas = input("Insira as notas: ")
    notas = notas.split() 
    notas1 = []
    for nota in notas:
        notas1.append(float(nota)) 
    aluno = (nome, id, notas1)
    turma.append(aluno)
    print("Aluno inserido com sucesso!")

def listar_turma(turma):
    if not turma:
        print("Não há alunas na turma.")
        return
    for aluno in turma:
        print(f"Nome: {aluno[0]}, ID: {aluno[1]}, Notas: {aluno[2]}")

def consultar_aluno(turma):
    id = input("Insira o ID do aluno a ser consultado: ")
    for aluno in turma:
        if aluno[1] == id:
            print(f"Nome: {aluno[0]}, ID: {aluno[1]}, Notas: {aluno[2]}")
            return
    print("Aluno não encontrado.")

def guardar_turma(turma):
    filename = input("Insira o nome do ficheiro para guardar a turma: ")
    with open(filename, 'w') as f:
        for aluno in turma:
            notas_str = ''
            for nota in aluno[2]:
                notas_str += str(nota) + ' ' 
            f.write(f"{aluno[0]},{aluno[1]},{notas_str.strip()}\n")  
    print("Turma guardada com sucesso!")

def carregar_turma():
    filename = input("Insira o nome do ficheiro para carregar a turma: ")
    turma = []
    
    f = open(filename, 'r')
    if f:
        for line in f:
            dados = line.strip().split(',')
            nome = dados[0]
            id = dados[1]
            notas = dados[2].split()
            notas_float = []
            for nota in notas:
                notas_float.append(float(nota))
            aluno = (nome, id, notas_float)
            turma.append(aluno)
        f.close()
        print("Turma carregada com sucesso!")
    else:
        print("Ficheiro não encontrado.")

    return turma

def menu():
    turma = criar_turma()
    while True:
        print("\nMenu de Operações:")
        print("1: Criar uma turma")
        print("2: Inserir um aluno na turma")
        print("3: Listar a turma")
        print("4: Consultar um aluno por id")
        print("8: Guardar a turma em ficheiro")
        print("9: Carregar uma turma a partir de um ficheiro")
        print("0: Sair da aplicação")
        
        opcao = input("Selecione uma opção: ")

        if opcao == '1':
            turma = criar_turma()
            print("Turma criada com sucesso!")
        elif opcao == '2':
            inserir_aluno(turma)
        elif opcao == '3':
            listar_turma(turma)
        elif opcao == '4':
            consultar_aluno(turma)
        elif opcao == '8':
            guardar_turma(turma)
        elif opcao == '9':
            turma = carregar_turma()
        elif opcao == '0':
            print("Saiu da aplicação")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    menu()
```
